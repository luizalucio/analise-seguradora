```python
## 1. Padrões de preços e custos por sexo
# Possíveis diferenças nos padrões de preços e nos 
# custos dos sinistros em relação ao sexo dos proprietários
```

```python
# frequência e indenização
colunas_frequencia = ["FREQ_SIN1", "FREQ_SIN2", "FREQ_SIN3", "FREQ_SIN4", "FREQ_SIN9"]
colunas_indenizacao = ["INDENIZ1", "INDENIZ2", "INDENIZ3", "INDENIZ4", "INDENIZ9"]

# agrupando 
agrupado = base_sp_filtrada.groupby("SEXO").agg({
    "EXPOSICAO1": "sum",
    "PREMIO1": "sum",
    **{col: "sum" for col in colunas_frequencia},  
    **{col: "sum" for col in colunas_indenizacao}  
}).reset_index()

# medianas diretas da base original
medianas = base_sp_filtrada.groupby("SEXO").agg(
    Mediana_Premio   = ("PREMIO1",  "median"),
    Mediana_Indeniz2 = ("INDENIZ2", "median"),
    Mediana_Indeniz3 = ("INDENIZ3", "median")
).reset_index()

# totais consolidados 
agrupado["TOTAL_FREQ"] = agrupado[colunas_frequencia].sum(axis=1)
agrupado["TOTAL_INDENIZ"] = agrupado[colunas_indenizacao].sum(axis=1)

# analises
analise_sexo = pd.DataFrame()
analise_sexo["Sexo"] = agrupado["SEXO"]

# Prêmio Médio
analise_sexo["Premio_Medio"] = agrupado["PREMIO1"] / agrupado["EXPOSICAO1"]

# Frequência Sinistral Total
analise_sexo["Freq_Sinistral_Total"] = agrupado["TOTAL_FREQ"] / agrupado["EXPOSICAO1"]

# Custo Médio por Exposto (Sinistro Médio)
analise_sexo["Custo_Medio_Exposto"] = agrupado["TOTAL_INDENIZ"] / agrupado["EXPOSICAO1"]

# Severidades Específicas
analise_sexo["Severidade_Roubo"] = agrupado["INDENIZ1"] / agrupado["FREQ_SIN1"]
analise_sexo["Severidade_Colisao_Parcial"] = agrupado["INDENIZ2"] / agrupado["FREQ_SIN2"]
analise_sexo["Severidade_Perda_Total"] = agrupado["INDENIZ3"] / agrupado["FREQ_SIN3"]

# Preenchendo possíveis valores nulos
analise_sexo = analise_sexo.fillna(0)

print(analise_sexo.to_string(index=False))
```

# TESTE T

```python
from scipy import stats
import numpy as np

# Lembrando que aqui se não tem essas bibliotecas é necessário instalar.

# excluindo J de jurídico

masculino = base_sp_filtrada[base_sp_filtrada["SEXO"] == "M"]
feminino  = base_sp_filtrada[base_sp_filtrada["SEXO"] == "F"]

def teste_t(var, nome):
    grupo_m = masculino[var].dropna()
    grupo_f = feminino[var].dropna()
    
    t_stat, p_valor = stats.ttest_ind(grupo_m, grupo_f, equal_var=False)
    
    media_m, media_f = grupo_m.mean(), grupo_f.mean()
    dp_pool = np.sqrt((grupo_m.std()**2 + grupo_f.std()**2) / 2)
    cohens_d = (media_m - media_f) / dp_pool
    
    print(nome)
    print("Média Masculino:", media_m)
    print("Média Feminino:", media_f)
    print("Estatística t:", t_stat)
    print("p-valor:", p_valor)
    
    # Interpretação automática do efeito
    d = abs(cohens_d)
    if d < 0.2:
        efeito = "negligenciável"
    elif d < 0.5:
        efeito = "pequeno"
    elif d < 0.8:
        efeito = "médio"
    else:
        efeito = "grande"
    
    if p_valor < 0.05:
        sig = "SIM"
    else:
        sig = "NÃO"
    
    print("Significativo (alfa=0.05):", sig, "| Tamanho do efeito:", efeito)
    print()

# Rodando para as variáveis de interesse
teste_t("PREMIO1",  "Prêmio médio (PREMIO1)")
teste_t("FREQ_SIN2", "Frequência de colisão (FREQ_SIN2)")
teste_t("INDENIZ2",  "Custo de indenização por colisão (INDENIZ2)")
```
